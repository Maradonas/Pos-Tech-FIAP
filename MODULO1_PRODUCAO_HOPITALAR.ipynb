Desafio: adicionar uma coluna com a regiao de cada estado

Desafio: adicione o seu estado nessa lista de 7 estados

```python
#df_usaveis.loc["Unidade da Federação"]
df_usaveis.index
```

```python
df_usaveis["Total"] = df_usaveis.sum(axis=1) #Eixo 1 soma nas colunas
df_usaveis.head()
```

```python
# Dicionário de mapeamento de códigos para regiões
regioes = {
    'Norte': [11, 12, 13, 14, 15, 16, 17],
    'Nordeste': [21, 22, 23, 24, 25, 26, 27, 28, 29],
    'Sudeste': [31, 32, 33, 35],
    'Sul': [41, 42, 43],
    'Centro-Oeste': [50, 51, 52, 53]
}

# Função para extrair os dois primeiros números do índice
def obter_codigo(estado):
    return int(str(estado).split()[0])  # Pega os primeiros números antes do espaço

# Aplicar função ao índice do DataFrame
df_usaveis['Código'] = df_usaveis.index.map(obter_codigo)

# Adicionar a coluna "Região" mapeando os códigos extraídos
df_usaveis['Região'] = df_usaveis['Código'].map(lambda x: next((regiao for regiao, estados in regioes.items() if x in estados), 'Desconhecido'))

# Remover a coluna auxiliar "Código" (opcional)
df_usaveis.drop(columns=['Código'], inplace=True)


```
